<h1><b><u>Bayesian Theorem Notebook</u></b></h1>

* Baye's theorem named after mathematician `Thomas Bayes`, is a way to `calculate
the probability of an event happening ` given that you already know something has happened. In other words it lets you update your initial guese(prior probability) about how likely something is to happen (event A) after you know about (event B)

<ol>
<li><b>Prior probability</b>- This is your initial guess of how likely event A is to happen, before you consider any new thing
<li><b>Conditional probability</b>-This is the likelihood of event A happening given that you know event B already happened. This is what Bayes' theorem helps you calculate.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. **Define the Probabilities**:
   - We set the initial probabilities for the medical test scenario. These include the prevalence of the disease in the population (`P(Disease)`) and the accuracy rates of the test (`P(Positive | Disease)` and `P(Positive | No Disease)`).

```python
# Define the probabilities
# Hypothetical scenario:
# P(Disease) = Probability of having the disease
# P(No Disease) = Probability of not having the disease
# P(Positive | Disease) = Probability of testing positive given the disease
# P(Positive | No Disease) = Probability of testing positive given no disease
```

```python
P_Disease = 0.01  # 1% of the population has the disease
P_NoDisease = 1 - P_Disease  # 99% of the population does not have the disease
P_Positive_given_Disease = 0.99  # 99% chance of testing positive if you have the disease
P_Positive_given_NoDisease = 0.05  # 5% chance of testing positive if you don't have the disease
```

2. **Calculate \( P(\text{Positive}) \)**:
   - We compute the overall probability of a positive test result using the law of total probability.

```python
# Calculate P(Positive)
P_Positive = (P_Positive_given_Disease * P_Disease) + (P_Positive_given_NoDisease * P_NoDisease)
```

3. **Apply Bayes' Theorem**:
   - We use Bayes' theorem to calculate the probability of having the disease given a positive test result:
   \[
   P(\text{Disease} | \text{Positive}) = \frac{P(\text{Positive} | \text{Disease}) \cdot P(\text{Disease})}{P(\text{Positive})}
   \]

```python
# Apply Bayes' Theorem to calculate P(Disease | Positive)
# P(Disease | Positive) = [P(Positive | Disease) * P(Disease)] / P(Positive)
P_Disease_given_Positive = (P_Positive_given_Disease * P_Disease) / P_Positive
```

4. **Display the Results**:
   - We print the probability of having the disease given a positive test result and create a DataFrame to summarize all the probabilities.


```python
#  Display the results
print("Probability of having the disease given a positive test result (P(Disease | Positive)):")
print(P_Disease_given_Positive)
```

5. **Visualization**:
   - We create a pie chart to visualize the distribution of disease prevalence in the population.
   - We create a bar chart to visualize the conditional probabilities of the test results.

```python
# Create a DataFrame to display the results
data = {
    'Event': ['P(Disease)', 'P(No Disease)', 'P(Positive | Disease)', 'P(Positive | No Disease)', 'P(Positive)', 'P(Disease | Positive)'],
    'Probability': [P_Disease, P_NoDisease, P_Positive_given_Disease, P_Positive_given_NoDisease, P_Positive, P_Disease_given_Positive]
}
```

```python
# Step 6: Visualization

# Pie chart of the probabilities
labels = ['Disease', 'No Disease']
sizes = [P_Disease, P_NoDisease]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # explode the 1st slice

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Probability of Disease vs No Disease in the Population')
plt.show()
```

```python
# Bar chart of conditional probabilities
labels = ['P(Positive | Disease)', 'P(Positive | No Disease)']
sizes = [P_Positive_given_Disease, P_Positive_given_NoDisease]

plt.figure(figsize=(8, 6))
plt.bar(labels, sizes, color=['lightcoral', 'lightskyblue'])
plt.ylabel('Probability')
plt.title('Conditional Probabilities of Test Results')
plt.show()
```
